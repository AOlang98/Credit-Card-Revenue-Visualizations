```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df_costs is your DataFrame
sns.lineplot(x='Year (cost)', y='CostPerType_Q', hue='Type (cost)', data=cost_df)
plt.title('Time Series Plot of CostPerType_Q Over Years')

# Save the image to the specified directory
plt.savefig(r'C:\\Users\\smrc\\OneDrive\\Desktop\\Credit Card Revenue Prediction\\images and plots\\time_series_plot1.png')

# Show the plot
plt.show()
```

```python
sns.barplot(x='Type (cost)', y='CostPerType_Q', data=cost_df, ci=None)
plt.title('Aggregated Costs Per Type')

# Save the image to the specified directory
plt.savefig(r'C:\\Users\\smrc\\OneDrive\\Desktop\\Credit Card Revenue Prediction\\images and plots\\time_series_plot2.png')

# Show the plot
plt.show()
```

```python
sns.boxplot(x='Type (cost)', y='CostPerType_Q', data=cost_df)
plt.title('Box Plot of Costs Per Type')

# Save the image to the specified directory
plt.savefig(r'C:\\Users\\smrc\\OneDrive\\Desktop\\Credit Card Revenue Prediction\\images and plots\\time_series_plot3.png')

# Show the plot
plt.show()
```
